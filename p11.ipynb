```python
def add(a,b): return a+b
def sub(a,b): return a-b
def mul(a,b): return a*b
def div(a,b): return a/b

a = float(input("a: "))
b = float(input("b: "))
op = input("op (+,-,*,/): ")

if op == "+": print(add(a,b))
elif op == "-": print(sub(a,b))
elif op == "*": print(mul(a,b))
elif op == "/":
    if b == 0: print("Divide by zero")
    else: print(div(a,b))
else: print("Invalid op")
```
