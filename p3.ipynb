```python
table = {
    ("A", "Clean"): "MoveRight",
    ("A", "Dirty"): "Suck",
    ("B", "Clean"): "MoveLeft",
    ("B", "Dirty"): "Suck",
}

percept = [("A", "Dirty"), ("A", "Clean"), ("B", "Dirty")]
agent_table = {}

def lookup(percept, table):
    action = table.get(percept)
    return action

for p in percept:
    action = lookup(p, table)
    agent_table[p] = action
print(agent_table)
```
