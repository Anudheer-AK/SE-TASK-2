```python
marks = list(map(int, input("Enter marks: ").split()))
avg = sum(marks) / len(marks)
if avg >= 90: g = "A"
elif avg >= 75: g = "B"
elif avg >= 60: g = "C"
else: g = "D"
print("Average =", avg, "Grade =", g)
```
