```python
bal = 1000
while True:
    print("\n1.Deposit 2.Withdraw 3.Balance 4.Exit")
    ch = input("Choice: ")
    if ch == "1":
        bal += int(input("Amount: "))
    elif ch == "2":
        amt = int(input("Amount: "))
        if amt <= bal: bal -= amt
        else: print("No money")
    elif ch == "3":
        print("Balance =", bal)
    elif ch == "4":
        break
    else:
        print("Invalid")
```
